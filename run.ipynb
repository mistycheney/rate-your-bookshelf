```python
%load_ext autoreload
%autoreload 2

import pandas as pd
from ryb import *
from tqdm import tqdm
from datetime import datetime
import logging
output_time = datetime.now().strftime('%Y%m%d%H%M%S')
logging.basicConfig(level=logging.INFO, handlers=[
        logging.FileHandler(f'app_{output_time}.log'),
        logging.StreamHandler()
     ])
pd.set_option('display.max_rows', None)
```

```python
image_dir = '/Users/yuncong/rate-your-bookshelf/uploads/'
output_dir = '/Users/yuncong/rate-your-bookshelf/output/'

for fp in sorted(os.listdir(image_dir)):
    img_path = os.path.join(image_dir, fp)

    logging.info(f'Processing image: {img_path}')
    books = process_image_with_gpt(img_path)
    logging.info(f'Found {len(books)} books in the image')
    books_df = pd.DataFrame(books)
    books_df['image_path'] = img_path

    output_time = datetime.now().strftime('%Y%m%d%H%M%S')
    books_df.to_csv(f'output/books_{output_time}.csv', index=False)
    logging.info(f'Saved output to books_{output_time}.csv')

    book_infos = []
    for _, book in books_df.iterrows():
        title = book.get('title')
        author = book.get('author')
        logging.info(f'Processing book: {title} by {author}')
        if title:
            book_info = get_goodreads_info(title, author)
            logging.info(f'Found book info: {book_info}')
            book_infos.append(book | book_info)
        # time.sleep(2)  # Be polite and don't hammer the server

    df = pd.DataFrame(book_infos)
    display(df.sort_values('rating_value', ascending=False))

    output_time = datetime.now().strftime('%Y%m%d%H%M%S')
    df.to_csv(f'output/ratings_{output_time}.csv', index=False)
```

```python
# load all csvs with ratings_* prefix and concatenate them
ratings_files = [f for f in os.listdir(output_dir) if f.startswith('ratings_')]
ratings_dfs = [pd.read_csv(os.path.join(output_dir, f)) for f in ratings_files]
ratings_df = pd.concat(ratings_dfs).reset_index(drop=True)
logging.info(f'Loaded {len(ratings_df)} ratings from {len(ratings_files)} files')

output_time = datetime.now().strftime('%Y%m%d%H%M%S')
ratings_df.to_csv(f'output/all_ratings_{output_time}.csv', index=False)
ratings_df.to_excel(f'output/all_ratings_{output_time}.xlsx', index=False)
display(ratings_df.sort_values('rating_value', ascending=False))
logging.info(f'Saved all ratings to output/all_ratings_{output_time}.csv')
```
